```python
import pandas as pd
import plotly.express as px
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load Data
raw_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DSCI521_Project/project/data/timeseries_data.csv', parse_dates=['date'])
raw_data.head()
```

```python
# subset to icu_bed_utilization
icu_df = raw_data[['date', 'state', 'adult_icu_bed_utilization']]
```

```python
icu_df['date'] = icu_df.loc[ :,'date'].dt.strftime('%Y-%m-%d')
```

```python
px.choropleth(icu_df,               
              locations="state",               
              color="adult_icu_bed_utilization",
              hover_name="state", 
              locationmode='USA-states',
              scope="usa",
              animation_frame="date",    
              color_continuous_scale='Plasma',  
              height=600             
)
```
