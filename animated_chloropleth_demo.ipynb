```python
import pandas as pd
import plotly.express as px
from plotly.offline import iplot
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load Data
raw_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_df.csv', parse_dates=['date'])
raw_data.head()
```

```python
# subset to icu_bed_utilization
icu_df = raw_data[['date', 'state', 'adult_icu_bed_utilization']]
icu_df = icu_df.sort_values(['date'], ascending=[True])
```

```python
icu_df_plot = icu_df
icu_df_plot['date'] = icu_df.loc[ :,'date'].dt.strftime('%Y-%m-%d')
```

```python
fig = px.choropleth(icu_df_plot,               
              locations="state",               
              color="adult_icu_bed_utilization",
              hover_name="state", 
              locationmode='USA-states',
              scope="usa",
              animation_frame="date",    
              color_continuous_scale='Inferno_r',  
              height=600             
)


fig.update_layout(
    title_text = 'Adult ICU Bed Utilization Across the US',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))

fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 75
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 10

iplot(fig)
```
